```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sir_hoax_simulation(beta, gamma, alpha, S0, I0, R0, H0, T=50, dt=0.1):
    n = int(T/dt)
    S = np.zeros(n)
    I = np.zeros(n)
    R = np.zeros(n)
    H = np.zeros(n)
    t = np.linspace(0, T, n)

    S[0], I[0], R[0], H[0] = S0, I0, R0, H0

    for k in range(n - 1):
        dS = -beta * S[k] * I[k]
        dI = beta * S[k] * I[k] - (gamma + alpha) * I[k]
        dR = gamma * I[k]
        dH = alpha * I[k]

        S[k+1] = S[k] + dS * dt
        I[k+1] = I[k] + dI * dt
        R[k+1] = R[k] + dR * dt
        H[k+1] = H[k] + dH * dt

    return t, S, I, R, H
```

```python
beta = 0.8     # penyebaran hoaks
gamma = 0.2    # edukasi
alpha = 0.05   # dampak hoaks

S0 = 0.95
I0 = 0.05
R0 = 0
H0 = 0

t, S, I, R, H = sir_hoax_simulation(beta, gamma, alpha, S0, I0, R0, H0)
```

```python
plt.figure(figsize=(10,5))
plt.plot(t, S, label='Susceptible')
plt.plot(t, I, label='Infected (Penyebar Hoaks)')
plt.plot(t, R, label='Recovered (Sadar Fakta)')
plt.plot(t, H, label='Harmed (Korban Hoaks)')
plt.xlabel('Waktu')
plt.ylabel('Proporsi Populasi')
plt.title('Model Penyebaran Hoaks - SIR Termodifikasi')
plt.grid(True)
plt.legend()
plt.show()
```

```python
R0_value = beta / (gamma + alpha)
print("Nilai R0 =", R0_value)

r = beta * S0 - (gamma + alpha)
print("Laju pertumbuhan awal =", r)

doubling_time = np.log(2)/r
print("Waktu penggandaan =", doubling_time)
```
